```python
import pandas as pd
import ast
import matplotlib.pyplot as plt
```

```python
data = {}
with open('model_1_metrics_extra.txt', 'r') as file:
    for line in file:
        key, value = line.split(':')
        data[key.strip()] = ast.literal_eval(value.strip())

df = pd.DataFrame(data)
```

```python
plt.plot(df['Mec ffnn'],df['Train accuracy'], marker='o', linestyle='-', color='r',label='Train accuracy')
plt.legend()
plt.savefig('train_accuracy.png')
plt.show()
```

```python
plt.plot(df['Mec ffnn'],df['Test accuracy'], marker='o', linestyle='-', color='r',label='Test accuracy')
plt.legend()
plt.savefig('test_accuracy.png')
plt.show()
```

```python
plt.plot(df['Mec ffnn'],df['Generalization'], marker='o', linestyle='-', color='r',label='Generalization')
plt.xlabel("MEC")
plt.ylabel("Generalization")
plt.legend()
plt.savefig('generalization.png')
plt.show()
```
